```python
import pandas as pd

df = pd.read_csv("NBAGamesBetween2017-2023Test.csv")
df
```

```python
# Df maÃ§ tarihine gÃ¶re sÄ±ralanÄ±yor
df = df.sort_values("GAME_DATE")
df = df.reset_index(drop=True)
```

```python
# Sezon sÃ¼tunu ekleniyor
df.SEASON_ID = df.SEASON_ID.astype(str)
df["SEASON"] = [season[-4:] for season in df["SEASON_ID"]]
```

```python
del df['MIN_A']
del df['MIN_B']
del df['TEAM_ID_A']
del df['TEAM_ID_B']
del df['TEAM_NAME_A']
del df['TEAM_NAME_B']
del df['PLUS_MINUS_A']
del df['PLUS_MINUS_B']

del df["SEASON_ID"]
del df["GAME_ID"]
del df['WL_B']

# Okunulabilirlik iÃ§in daha sonra silinecek 
# del df['MATCHUP_A']
# del df['MATCHUP_B']
```

```python
# Modelin Ã¶ÄŸrenmesi iÃ§in target sÃ¼tunu oluÅŸturuluyor.

def add_target(team):
    team["TARGET"] = team["WL_A"].shift(-1)
    return team

df = df.groupby("TEAM_ABBREVIATION_A", group_keys=False).apply(add_target)
```

```python
# Target sÃ¼tunu int deÄŸerlere dÃ¶nÃ¼ÅŸtÃ¼rÃ¼lÃ¼yor

df["TARGET"][pd.isnull(df["TARGET"])] = 2
df["TARGET"] = df["TARGET"].map({'W': 1 , "L":0, 2:2})
```

# MODELÄ°N EÄžÄ°TÄ°LMESÄ°


```python
from sklearn.model_selection import TimeSeriesSplit
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression


model = LogisticRegression()
#DatayÄ± tarihe gÃ¶re bÃ¶lmek iÃ§in time series split
split = TimeSeriesSplit(n_splits=3)
sfs = SequentialFeatureSelector(model, n_features_to_select=30, direction="forward", cv=split,n_jobs=1)
```

```python
removed_columns = ["SEASON", "GAME_DATE","WL_A","TARGET","TEAM_ABBREVIATION_A","TEAM_ABBREVIATION_B"]
selected_columns = df.columns[~df.columns.isin(removed_columns)]
```

```python
scaler = MinMaxScaler()
df[selected_columns] = scaler.fit_transform(df[selected_columns])
```

```python
df
```

```python
sfs.fit(df[selected_columns], df["TARGET"])
```

```python
predictors = list(selected_columns[sfs.get_support()])
predictors
```

```python
sorted(df["SEASON"].unique())
```

```python
def backtest(data, fun_model, predictors, start=2):
    all_predictions = []
    seasons = sorted(data["SEASON"].unique())

    for i in range(start, len(seasons)):
        season = seasons[i]
        train = data[data["SEASON"] < season]
        test = data[data["SEASON"] == season]
        
        fun_model.fit(train[predictors], train["TARGET"])
        predictions = model.predict(test[predictors])
        predictions = pd.Series(predictions, index=test.index)

        combined = pd.concat([test["TARGET"], predictions], axis=1)
        combined.columns = ["real", "prediction"]

        all_predictions.append(combined)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(df, model, predictors)
predictions
```

```python
from sklearn.metrics import accuracy_score

predictions = predictions[predictions["real"] != 2]
accuracy_score(predictions["real"], predictions["prediction"])
```
